```python
import pandas as pd
import os
import csv
```

```python
cbData = "data/2018/201810-citibike-tripdata.csv"
df = pd.read_csv(cbData)
df.head()
```

```python
df = df[pd.notnull(df['start station id'])]
```

```python
df['end station id'].unique()
```

```python
df.count()
```

```python
df = df.astype({"start station id": int, "end station id": int})
df.head()
```

```python
start_df = df.loc[:,['tripduration','starttime','stoptime','start station id','start station latitude','start station longitude','bikeid', 'usertype', 'birth year', 'gender']]
start_df['trip type'] = 'Start'
start_df.columns = ['tripduration','starttime','stoptime','station id','latitude','longitude','bikeid', 'usertype', 'birth year', 'gender','trip type']
start_df.head()
```

```python
end_df = df.loc[:,['tripduration','starttime','stoptime','end station id','end station latitude','end station longitude','bikeid', 'usertype', 'birth year', 'gender']]
end_df['trip type'] = 'End'
end_df.columns = ['tripduration','starttime','stoptime','station id','latitude','longitude','bikeid', 'usertype', 'birth year', 'gender','trip type']
end_df.head()
```

```python
frames = [start_df, end_df]
df_combined = pd.concat(frames, axis=0, join='inner')
df_combined['station id'] = df_combined['station id'].astype(str)
df_combined.dtypes
```

```python
df_combined.count()
```

```python
df_combined.to_csv('data/cb_data_2018_10_combined.csv', sep=',', encoding='utf-8', index=False)
```
