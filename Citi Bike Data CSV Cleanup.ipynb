```python
file = 'data/2018/201810-citibike-tripdata.csv'

df = pd.read_csv(file)
df.head()
```

```python
df = df[pd.notnull(df['start station id'])]
df.count()
```

```python
df = df.astype({"start station id": int, "end station id": int})
df.head()
```

```python
df.to_csv('data/cb_data_2018_10_clean.csv', sep=',', encoding='utf-8', index=False)
```
